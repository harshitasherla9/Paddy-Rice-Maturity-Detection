```python
import openpyxl
import torch
import pandas as pd
# Create a new workbook and worksheet
workbook = openpyxl.Workbook()
worksheet = workbook.active

# Write headers
headers = ["Image", "Class", "Confidence"]
worksheet.append(headers)
classes={0: 'initiation', 1: 'mature', 2: 'nursery', 3: 'transplanting', 4: 'visible'}
print(classes)
for result in results:
    # Get the bounding box coordinates and class predictions
    boxes = result.boxes

    # Iterate through the boxes
    for box in boxes:
        # Get the class index and confidence value
        class_index = box.cls
        class_index = int(class_index)
        confidence = box.conf
        print(class_index, confidence)
        # Append the data to the worksheet
        data = [img_path, classes[class_index], float(confidence)]
        worksheet.append(data)
    print()
# Save the Excel file
workbook.save("bbox_confidences.xlsx")
```
